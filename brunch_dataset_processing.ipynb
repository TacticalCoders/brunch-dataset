```python
import pandas as pd
```

```python
df_it = pd.read_csv('IT_dataset.csv')
df_health = pd.read_csv('건강_dataset.csv')
df_reading = pd.read_csv('독서_dataset.csv')
df_love = pd.read_csv('사랑_dataset.csv')
df_mental = pd.read_csv('심리_dataset.csv')
df_essay = pd.read_csv('에세이_dataset.csv')
df_travel = pd.read_csv('여행_dataset.csv')
df_history = pd.read_csv('역사_dataset.csv')
df_movie = pd.read_csv('영화_dataset.csv')
df_art = pd.read_csv('예술_dataset.csv')
df_exercise = pd.read_csv('운동_dataset.csv')
df_music = pd.read_csv('음악_dataset.csv')
df_career = pd.read_csv('직장생활_dataset.csv')
df_foundation = pd.read_csv('창업_dataset.csv')
df_philosophy = pd.read_csv('철학_dataset.csv')
df_parenting = pd.read_csv('육아_dataset.csv')
df_humanities = pd.read_csv('인문학_dataset.csv')
df_politic = pd.read_csv('정치_dataset.csv')
df_photo = pd.read_csv('사진_dataset.csv')
df_farewell = pd.read_csv('이별_dataset.csv')
df_write = pd.read_csv('글쓰기_dataset.csv')
df_stock = pd.read_csv('주식_dataset.csv')
df_property = pd.read_csv('부동산_dataset.csv')
```

```python
df_list = [df_it, df_health, df_reading, df_love, 
           df_mental, df_essay, df_travel, df_history,
           df_movie, df_art, df_exercise, df_music,
           df_career, df_foundation, df_philosophy,
           df_parenting, df_humanities, df_politic,
           df_photo, df_farewell, df_write, df_stock,
           df_property]
```

```python
pd.set_option('display.max.colwidth', 200)
```

```python
for df in df_list:
    display(df[['title', 'body_text', 'keyword']])
```

```python
# '\r' 제거
for df in df_list:
    df['body_text'] = df['body_text'].str.replace('\r', " ")
    display(df[['body_text']])
```

```python
# https or http~ 형태의 url 제거
for df in df_list:
    df['body_text'] = df['body_text'].str.replace(pat=r"((http|https)\:\/\/)?[a-zA-Z0-9\.\/\?\:@\-_=#]+\.([a-zA-Z]){2,6}([a-zA-Z0-9\.\&\/\?\:@\-_=#%])*", 
                                                 repl=r' ', regex=True)
    display(df[['body_text']])
```

```python
# %를 '퍼센트'로 변경
for df in df_list:
    df['title'] = df['title'].str.replace(pat=r"%", repl=r'퍼센트', regex=True)
    df['sub_title'] = df['sub_title'].str.replace(pat=r"%", repl=r'퍼센트', regex=True)
    df['body_text'] = df['body_text'].str.replace(pat=r"%", repl=r'퍼센트', regex=True)
    df['author_desc'] = df['author_desc'].str.replace(pat=r"%", repl=r'퍼센트', regex=True)
```

```python
# 콤마제거
for df in df_list:
    df['title'] = df['title'].str.replace(pat=r",", repl=r'', regex=True)
    df['sub_title'] = df['sub_title'].str.replace(pat=r",", repl=r'', regex=True)
    df['body_text'] = df['body_text'].str.replace(pat=r",", repl=r'', regex=True)
    df['author_desc'] = df['author_desc'].str.replace(pat=r",", repl=r'', regex=True)
    display(df[['body_text']])
```

```python
# .을 제외한 특수문자 제거
for df in df_list:
    df['title'] = df['title'].str.replace(pat=r"[^0-9a-zA-Zㄱ-ㅎ가-힣.]", repl=r' ', regex=True)
    df['sub_title'] = df['sub_title'].str.replace(pat=r"[^0-9a-zA-Zㄱ-ㅎ가-힣.]", repl=r' ', regex=True)
    df['body_text'] = df['body_text'].str.replace(pat=r"[^0-9a-zA-Zㄱ-ㅎ가-힣.]", repl=r' ', regex=True)
    df['author_desc'] = df['author_desc'].str.replace(pat=r"[^0-9a-zA-Zㄱ-ㅎ가-힣.]", repl=r' ', regex=True)
    display(df[['body_text']])
```

```python
# 두 개 이상의 . 제거
for df in df_list:
    df['title'] = df['title'].str.replace(pat=r"\.+", repl=r'.', regex=True)
    df['sub_title'] = df['sub_title'].str.replace(pat=r"\.+", repl=r'.', regex=True)
    df['body_text'] = df['body_text'].str.replace(pat=r"\.+", repl=r'.', regex=True)
    df['author_desc'] = df['author_desc'].str.replace(pat=r"\.+", repl=r'.', regex=True)
    display(df[['title','body_text']])
```

```python
# 본문이 500자 미만 행 제거
cnt_total = 0
for df in df_list:
    print('제거 전:', df.shape[0])
    idx = df['body_text'][df['body_text'].str.len() < 500].index
    df.drop(idx, inplace=True)
    print('제거 후:', df.shape[0], end='\n\n')
    cnt_total += df.shape[0]
```

```python
print('total:', cnt_total)
```

```python
# 중복 공백 제거
for df in df_list:
    df['title'] = df['title'].str.replace(pat=r" +", repl=r' ', regex=True)
    df['sub_title'] = df['sub_title'].str.replace(pat=r" +", repl=r' ', regex=True)
    df['body_text'] = df['body_text'].str.replace(pat=r" +", repl=r' ', regex=True)
    df['author_desc'] = df['author_desc'].str.replace(pat=r" +", repl=r' ', regex=True)
```

```python
# 인덱스 초기화
for df in df_list:
    df.reset_index(inplace=True, drop=True)
```

```python
for df in df_list:
    display(df[['title','body_text', 'keyword', 'post_date']])
```
